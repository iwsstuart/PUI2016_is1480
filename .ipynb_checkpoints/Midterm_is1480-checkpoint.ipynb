```python
# Define a simple function to retrieve data for a specific month and save it in PUIdata
def getCitiBikeCSV(datestring):
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):    
        os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.zip")
        os.system("mv " + datestring + "-citibike-tripdata.zip " + os.getenv("PUIDATA"))
        os.system("unzip " + os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip")
        os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA"))
```

```python
# Loop through all the latlon pairs and make call to geocoding api to return address
# Extract zipcode from json that is returned, and insert 
# into zipcode column of citibike data where latlon matches

for latlon in pd.DataFrame(station_latlon).drop_duplicates().values:
    url = ("https://maps.googleapis.com/maps/api/geocode/json?latlng=" +
           "%f,%f&key=%s"%(
            latlon[0], latlon[1], apikey))
#     print (latlon, url)
#     print (get_jsonparsed_data(url))
    revgeo = get_jsonparsed_data(url)["results"][0]['address_components'][-1]['long_name']
#     print (revgeo)
    june15['zipcodes'][(june15['start station latitude'] == latlon[0]) * 
           (june15['start station longitude'] == latlon[1])] = revgeo
    
june15.head()
```

```python
# group data by zipcode and get counts of trips for each zipcode

cbgroup = june15.groupby('zipcodes', as_index=False).count()
cbgroup.head()
```

```python
# Convert zipcodes to numeric and drop those less than 10000
cbgroup['zipcodes'] = pd.to_numeric(cbgroup['zipcodes'], errors='coerce')

cbgroup = cbgroup[cbgroup['zipcodes'] > 9999]
cbgroup.head()
```

```python
# Drop columns and rename counts
cbgroup = cbgroup.iloc[:,:2]
cbgroup.rename(columns={'tripduration':'Nrides'}, 
           inplace=True)

# Merge datasets on zipcodes
cbincome = pd.merge(cbgroup, zipincome3, on="zipcodes")
cbincome.head()
```

```python
x = cbincome.income
y = cbincome.Nrides

lm = sm.OLS(y, sm.add_constant(x)).fit()
lm.summary()
```

```python
# Save the coefficient and intercept
beta0_sm, beta1_sm = lm.params

pl.figure(figsize=(8,6))
pl.plot(x, x * beta1_sm + beta0_sm, 'r', label = 'fitted line')
pl.plot(x, y, 'o')

pl.xlabel('Zip Code AGI', fontsize=14)
pl.ylabel('Number of Rides from Zip Code', fontsize=14)
pl.title('Linear Regression - Ridership by Zipcode vs AGI (June 2015)')
pl.legend()
```

#### Figure 1: Adjusted Gross Income (AGI) plotted against the number of CitiBike rides originating in each zip code, with a fitted line obtained by linear regression. The regression indicates that ridership increases with AGI.

```python
# Remove two highest income data points
cbincome_cut = cbincome[cbincome['income'] < 1.2e7]
```

```python
# Redo the regression

x = cbincome_cut.income
y = cbincome_cut.Nrides

lm = sm.OLS(y, sm.add_constant(x)).fit()
lm.summary()
```

```python
pl.figure(figsize=(8,6))
pl.plot(x, x * beta1_sm + beta0_sm, 'r', label = 'fitted line')
pl.plot(x, y, 'o')

pl.xlabel('Zip Code AGI', fontsize=14)
pl.ylabel('Number of Rides from Zip Code', fontsize=14)
pl.title('Linear Regression - Ridership by Zipcode vs AGI (June 2015)')
pl.legend(loc='upper left')
```

#### Figure 2: Adjusted Gross Income (AGI) plotted against the number of CitiBike rides originating in each zip code, with a fitted line obtained by linear regression. Top two income values removed.

```python
modeldf = pd.DataFrame()

modeldf['x'] = cbincome_cut.income
modeldf['x2'] = cbincome_cut.income * cbincome_cut.income
modeldf['y'] = cbincome_cut.Nrides

curvemodel = smf.ols(formula = 'y ~ x2 + x', 
                          data = modeldf).fit()

curvemodel.summary()
```

```python
# Create new dataframe 'x' and 'x2' values to feed to model
# Use enough points to make curve appear smooth

predictdf = pd.DataFrame()
predictdf['x'] = np.linspace(cbincome_cut['income'].min(), cbincome_cut['income'].max(), 1000)
predictdf['x2'] = predictdf['x'] * predictdf['x']
```

```python
pl.figure(figsize=(8,6))
pl.plot(x, y, 'o')
pl.plot(predictdf['x'], curvemodel.predict(predictdf), 'r', label = '2nd Degree Fit')

pl.xlabel('Zip Code AGI', fontsize=14)
pl.ylabel('Number of Rides from Zip Code', fontsize=14)
pl.title('2nd Degree Polynomial Fit - Ridership by Zipcode vs AGI (June 2015)')
pl.legend(loc='upper left')
```

#### Figure 3: Adjusted Gross Income (AGI) plotted against the number of CitiBike rides originating in each zip code, with a 2nd degree polynomial fit. Top two income values removed.

```python
# Likelihood ratio test to compare the linear and 2nd degree models

alpha = 0.05
print ("LR from statsmodels:", curvemodel.compare_lr_test(lm))
LR = curvemodel.compare_lr_test(lm)

print("LR statistic = ", + LR[0])

print ("We ", end="")
if LR[0] < 3.84: #0.05 level for 1 DOF chi sq distribution 
    print ("CANNOT ", end="") 
    
print ("reject the Null hypothesis that the linear " + 
       "model is better than the 2nd degree polynomial model with p-value ", end="")
print ("p < %.3f"%alpha)
```

```python
# Likelihood ratio test to compare the linear and 2nd degree models for Jan 2016

alpha = 0.05
print ("LR from statsmodels:", curvemodel_2.compare_lr_test(lm2))
LR = curvemodel_2.compare_lr_test(lm2)

print("LR statistic = ", + LR[0])

print ("We ", end="")
if LR[0] < 3.84: #0.05 level for 1 DOF chi sq distribution 
    print ("CANNOT ", end="") 
    
print ("reject the Null hypothesis that the linear " + 
       "model is better than the 2nd degree polynomial model with p-value ", end="")
print ("p < %.3f"%alpha)
```

### In the January 2016 data, we see that the effect of the highest income zipcodes was larger, because when we remove those values and run the liklihood ratio test, the LR stat is higher than before (although not high enough to accept the 2nd degree model over the linear one).
