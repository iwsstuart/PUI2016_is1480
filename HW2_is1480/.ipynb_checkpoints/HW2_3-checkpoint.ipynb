```python
import os

print os.getenv("DFDATA")
```

```python
import pandas as pd
```

```python
dfData = os.getenv("DFDATA")

# location of building oil usage data in data facility
dataLoc = dfData + "/uedp-fegm/1414245967/uedp-fegm"

data = pd.read_csv(dataLoc)
print data.head()

print data.columns
```

```python
dataCols = data[[' Consumption (therms) ', ' Consumption (GJ) ']]
print dataCols.head()
```

```python
import pylab as pl
%pylab inline

pl.scatter(dataCols[' Consumption (therms) '], dataCols[' Consumption (GJ) '], facecolors="none", alpha=0.3)
pl.legend().get_texts()[0].set_text('Individual Buildings')
pl.xlabel("Consumption (therms)")
pl.ylabel("Consumption (GJ)")
pl.title('Building Gas Consumption in Therms vs. GJ')
```
